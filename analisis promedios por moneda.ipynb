```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('/content/sample_data/Base de datos usada global_food_prices.csv')

# Identify the most expensive and cheapest food products on average
average_prices = df.groupby('cm_name')['mp_price'].mean().sort_values()
cheapest_product = average_prices.idxmin()
most_expensive_product = average_prices.idxmax()

print(f'The cheapest product on average is: {cheapest_product}')
print(f'The most expensive product on average is: {most_expensive_product}')

# Analyze the relationship between the type of currency and food prices
currency_prices = df.groupby('cur_name')['mp_price'].mean().sort_values()

# Plot the average price for each currency
plt.figure(figsize=(10, 6))
sns.barplot(x=currency_prices.index, y=currency_prices.values)
plt.title('Average Price for Each Currency')
plt.xlabel('Currency')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
plt.show()
```
